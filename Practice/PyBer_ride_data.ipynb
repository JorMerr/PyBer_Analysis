```python
%matplotlib inline
# import dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
# Load in csv
pyber_ride_df = pd.read_csv("Resources/PyBer_ride_data.csv")
pyber_ride_df
```

```python
# set x-axis and tick locations.
x_axis = np.arange(len(pyber_ride_df))
tick_locations = [value for value in x_axis]
# Plot the data.
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)", color="skyblue")
plt.xticks(tick_locations, pyber_ride_df["Month"])
plt.show()
```

```python
pyber_ride_df.plot.bar(x="Month", y="Avg. Fare ($USD)")
plt.show()
```

```python
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)", kind="bar")
plt.show()
```

```python
import statistics
# Get the standard deviation of the values in the y-axis.
stdev = statistics.stdev(pyber_ride_df["Avg. Fare ($USD)"])
pyber_ride_df.plot.bar(x="Month", y="Avg. Fare ($USD)", color="skyblue", yerr=stdev , capsize=3, width=0.5, align='center')
plt.yticks(np.arange(0, 50, step=5.0))
plt.xticks(rotation=0)
plt.show()
```
